```python
text='''BARNES RICHARD C. BARNES Richard C. 'Rick' Barnes, 82, passed away, Sunday, May 22, 2022, at home in Alexandria, VA. He graduated from the University of Texas at Austin and later moved to Washington, DC where he worked as a staff member on the House Banking Committee and as Director of the Legislation and National Security Subcommittee. After retiring, Rick helped found the Independent Bank in Manassas,VA, and was a founding partner of Independent Associates. He is survived by his wife of 55 years, Sharon; two sons, Jeff of San Luis Obispo, CA, and Tim of Alexandria, VA; his brother Ben Barnes of Austin, Texas; and a nephew and nieces. A celebration of life will be June 24, 2 p.m., in the Oakbrooke Common Area, entrance just off Oakbrooke Avenue, Alexandria, VA 22308. Remembrances can be made to UT MD Anderson Cancer Center&#x27;s Department of Cardiology, P.O. Box 4486 Houston, Texas 77210-4486, or mdanderson.org/gifts. https://www.cunninghamfuneralhome.net/obituary/richard-c-barnes-rick/'''

txt=sent_tokenize(text)
print(txt)
```

```python
reqlist=['survived','surviving','survivor','Survivor','pecial friend','partner','leaves to ','leaves behind','left to','left behind','loved','beloved','missed','legacy','husband', 'wife', 'son','sons', 'daughter', 'brother', 'sister', 'mother', 'father', 'grandfather', 'grandmother']
startindex=[]
flag=0
l01=[]
l02=[]

for j in range(len(txt)):
  wt=word_tokenize(txt[j])
  for r in reqlist:
    for index,item in enumerate(wt):
      if(item==r):
        startindex.append(index)  
        list2=wt[index+1:]
        # print(list2)
        list3=' '.join(list2)
        l01.append(list3)

# print(l01)
          
for j in range(len(l01)):
  flag=0
  wt1=word_tokenize(l01[j])
  print(wt1)
  for index2, item2 in enumerate(wt1):
    for r2 in reqlist:
      if r2==item2:
        # print(item2)
        flag=1
    if flag==1:
      break;
    l02.append(item2)
print(l02)
```

```python
final=' '.join(l02)
print(final)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner1=ner(final)
print(ner1)
```
